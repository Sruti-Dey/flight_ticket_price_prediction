```python
def convert_duration(duration):
    if len(duration.split())==2:
        hourse=int(duration.split()[0][:-1])
        minutes=int(duration.split()[1][:-1])
        return hourse*60+minutes
    else:
        return int(duration[0:-1])*60
```

## Building Machine Learning Model

```python
temp_col=data.columns.to_list()
print(temp_col,'\n')

new_col=temp_col[:2]+temp_col[3:]
new_col.append(temp_col[2])
print(new_col,'\n')

data=data.reindex(columns=new_col)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
data=scaler.fit_transform(data)

data[0]
```

```python
from sklearn.model_selection import train_test_split as tts

x=data[:,: -1]
y=data[:,-1]
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.1,random_state=60)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error,r2_score

def metrics(y_true, y_predict):
    print(f'RMSE: ',mean_squared_error(y_true,y_pred)**0.5)
    print(f'R_squared value: ',r2_score(y_true,y_pred))
    
def accuracy(y_true,y_pred):
    errors=abs(y_true -y_pred)
    mape=100*np.mean(errors/y_true)
    accuracy=100-mape
    return accuracy
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
metrics(y_test,y_pred)
```

```python
accuracy(y_test,y_pred)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

model_rf=RandomForestRegressor(n_estimators=500, min_samples_split=3)
model_rf.fit(x_train,y_train)
```

```python
pred_rf=model_rf.predict(x_test)
```

```python
metrics(y_test,pred_rf)
```

```python
accuracy(y_test,pred_rf)
```
